### There are no null values

```python
for i in df.columns:
    print(i,'->',df[i].unique())
```

### veil-type has normal one unique value, so dropping it

```python
df.drop('veil-type', axis=1, inplace=True)
```

### All feature are object type, we are converting it into integer type by doing Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
```

```python
encoder
```

```python
for i in df.columns:
    df[i]=encoder.fit_transform(df[i])
```

```python
df.head()
```

```python
for i in df.columns:
    print(i,'->',df[i].unique())
```

```python
df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
corr_matrix = df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f',linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
corr_matrix
```

```python
### Gill color has high covariacne, so we can say that it is an important feature
```

```python
X=df.drop('class',axis=1)
y=df['class']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=10)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
```

```python
X_train_scaled=scalar.fit_transform(X_train)
X_test_scaled=scalar.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

# Model Training

### Multilinear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
regression
```

```python
regression.fit(X_train_scaled,y_train)
```

```python
## slopes of feature
regression.coef_
```

```python
regression.intercept_
```

```python
y_pred_test=regression.predict(X_test_scaled)
```

```python
## Performance metrics cost function
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print(mean_squared_error(y_test,y_pred_test))
print(mean_absolute_error(y_test,y_pred_test))
print(np.sqrt(mean_squared_error(y_test,y_pred_test)))
```

```python
## R square and Adjusted R square
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred_test)
print(score)
```

### LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression(max_iter=10000)
classifier.fit(X_train_scaled,y_train)
```

```python
## Prediction
y_pred=classifier.predict(X_test_scaled)
```

```python
## Performance metrics cost function
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report

print(confusion_matrix(y_pred,y_test))
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

### DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
## Post Pruning 
treeclassifier=DecisionTreeClassifier()
treeclassifier.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled=True)
```

```python
## Post Pruning  with max_depth=2
treeclassifier=DecisionTreeClassifier(max_depth=2)
treeclassifier.fit(X_train,y_train)
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled=True)
```

```python
## Prediction 
y_pred=treeclassifier.predict(X_test)
```

```python
score=accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
```

```python
## Post Pruning  with max_depth=2
treeclassifier=DecisionTreeClassifier(max_depth=2)
treeclassifier.fit(X_train,y_train)
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled=True)
```

```python
## Prediction 
y_pred=treeclassifier.predict(X_test)
```

```python
score=accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
```

### SVC

```python
from sklearn.svm import SVC
svc=SVC(kernel='linear')
svc.fit(X_train,y_train)
print(svc.coef_)
y_pred=svc.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```
